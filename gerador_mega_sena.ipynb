### Salvando html dos resultados Mega Sena

```python
# salvando dataframe via html
url = 'https://servicebus2.caixa.gov.br/portaldeloterias/api/resultados?modalidade=Mega-Sena'
resultados = pd.read_html(url, encoding='utf-8')[0]
resultados.head(2)
```

**Limpando dados dos resultados**

```python
# limpando \r\n dos dados
resultados = resultados.replace(r'\\r\\n','', regex=True)
resultados.columns = [col.replace(r'\r\n','') for col in resultados.columns]

# limpando linhas que não são resultados de concurso
resultados = resultados[resultados['Concurso'].str.contains(r'[0-9]{1,4}', na=False)]

# salvando em um excel
resultados.to_excel('resultados_mega_sena.xlsx', index=False)

resultados.head()
```

### Funções

```python
def display_side_by_side_space(dfs:list, captions:list):
    """Display tables side by side to save vertical space
    Input:
        dfs: list of pandas.DataFrame
        captions: list of table captions
    """
    output = ""
    combined = dict(zip(captions, dfs))
    for caption, df in combined.items():
        output += df.style.set_table_attributes("style='display:inline'").set_caption(caption)._repr_html_()
        output += "\xa0\xa0\xa0\xa0\xa0\xa0"
    display(HTML(output))
```

#### Lendo os dados dos resultados anteriores

```python
todos_resultados = resultados.copy()
todos_resultados = todos_resultados.set_index('Concurso')
todos_resultados = todos_resultados[todos_resultados.columns[1:7]]
todos_resultados.head()
```

#### Escolhendo 20 números

**Escolha de 20 números**

A escolha se dará da sequinte forma: <br>
1º - escolher 2 números que mais saíram nos últimos 100 jogos<br>
2º - escolher o número que mais saiu de cada quadrante principal<br>
3º - adicionar todos esses números em uma lista e remover os repetidos<br>
4º - escolher mini quadrante de cada um desses números<br>
5º - escolher 3 números de cada um desses mini quadrantes incluindo os que estão na lista<br>
6º - verificar se já tem 6 quadrantes preenchidos, caso não escolher x mini quadrantes para o que falta de 6 e selecionar 3 números de cada<br>
7º - tirando os mini quadrantes já selecionados, escolher um novo mini quadrante e escolher apenas 2 números dele

```python
todos_numeros = todos_resultados['Coluna1'].append(todos_resultados['Coluna2'])
todos_numeros = todos_numeros.append(todos_resultados['Coluna3'])
todos_numeros = todos_numeros.append(todos_resultados['Coluna4'])
todos_numeros = todos_numeros.append(todos_resultados['Coluna5'])
todos_numeros = todos_numeros.append(todos_resultados['Coluna6'])
todos_numeros.value_counts().head()
```

```python
# escolher 2 números que mais saíram nos 100 últimos jogos
dois_mais_apareceram = list(todos_numeros.tail(100).value_counts().head(2).index)
vinte_dezenas = dois_mais_apareceram

vinte_dezenas = [int(vinte_dezenas[n]) for n in range(0, len(vinte_dezenas))]
vinte_dezenas                
```

```python
# escolher o número que mais saiu de cada quadrante principal e
# adicionar todos esses números em uma lista
v1, v2, v3, v4 = False, False, False, False
for num in todos_numeros.value_counts().index:
    num = int(num)
    if (v1 and v2 and v3 and v4):
        break
    if not v1:
        if num in quad1.values:
            vinte_dezenas.append(num)
            v1 = True
    if not v2:
        if num in quad2.values:
            vinte_dezenas.append(num)
            v2 = True
    if not v3:
        if num in quad3.values:
            vinte_dezenas.append(num)
            v3 = True
    if not v4:
        if num in quad4.values:
            vinte_dezenas.append(num)
            v4 = True
vinte_dezenas
```

```python
# remover os repetidos
vinte_dezenas = sorted(set(vinte_dezenas))
vinte_dezenas
```

```python
# escolher mini quadrante de cada um desses números

# criar um dicionário para saber se quadrante foi selecionado e
# quantos números tem preenchidos

mini_quadrantes = {}
mini_quad = []
for i in range(1, 16):

    if i == 1:
        mini_quad = mini_quad1
    elif i == 2:
        mini_quad = mini_quad2
    elif i == 3:
        mini_quad = mini_quad3
    elif i == 4:
        mini_quad = mini_quad4
    elif i == 5:
        mini_quad = mini_quad5
    elif i == 6:
        mini_quad = mini_quad6
    elif i == 7:
        mini_quad = mini_quad7
    elif i == 8:
        mini_quad = mini_quad8
    elif i == 9:
        mini_quad = mini_quad9
    elif i == 10:
        mini_quad = mini_quad10
    elif i == 11:
        mini_quad = mini_quad11
    elif i == 12:
        mini_quad = mini_quad12
    elif i == 13:
        mini_quad = mini_quad13
    elif i == 14:
        mini_quad = mini_quad14
    elif i == 15:
        mini_quad = mini_quad15

    item = {f'mini_quad{i}': {'selecionado': False,
                               'quantidade': 0,
                               'dados': mini_quad}}
    mini_quadrantes.update(item)
```

```python
for num in vinte_dezenas:
    for i in range(1,16):
        
        df = mini_quadrantes[f'mini_quad{i}']['dados']
        if num in df.values:
            mini_quadrantes[f'mini_quad{i}']['quantidade'] += 1
            mini_quadrantes[f'mini_quad{i}']['selecionado'] = True
```

```python
# escolher 3 números de cada um desses mini quadrantes incluindo os que estão na lista
for qd in mini_quadrantes:
    selecao_tres = []
    if mini_quadrantes[qd]['selecionado']:
        values = (mini_quadrantes[qd]['dados']).values
        values = np.unique(values)
        
        num_intersection = list(set(values).intersection(vinte_dezenas))[0]
        selecao_tres.append(num_intersection)
        
        while len(selecao_tres) < 3:    
            num = random.choice(values) 
            if num not in selecao_tres:
                selecao_tres.append(num)
                
        for i in range(1,16):
            df = mini_quadrantes[f'mini_quad{i}']['dados']
            if num_intersection in df.values:
                mini_quadrantes[f'mini_quad{i}']['quantidade'] = 3
        
        vinte_dezenas.extend(selecao_tres[1:])

vinte_dezenas
```

```python
# 6º - verificar se já tem 6 quadrantes preenchidos,
# caso não escolher x mini quadrantes para o que falta de 6
# e selecionar 3 números de cada

count = 0
for i in range(1, 16):
    if mini_quadrantes[f'mini_quad{i}']['selecionado']:
        count += 1

nao_selecionados = []

for mini_quad in mini_quadrantes:
    if not mini_quadrantes[mini_quad]['selecionado']:
        nao_selecionados.append(mini_quad)

# escolher x mini quadrantes para o que falta
qtd_falta = 6 - count
if (qtd_falta) > 0:
    mini_quad = random.sample(nao_selecionados, qtd_falta)

    for i in range(0, qtd_falta):
        values = mini_quadrantes[mini_quad[i]]['dados'].values
        mini_quadrantes[mini_quad[i]]['quantidade'] = 3
        mini_quadrantes[mini_quad[i]]['selecionado'] = True
        values = np.unique(values)
        
        vinte_dezenas.extend(random.sample(list(values), 3))
vinte_dezenas
```

```python
#7º - tirando os mini quadrantes já selecionados, 
# escolher um novo mini quadrante e escolher apenas 2 números dele

nao_selecionados = []

for mini_quad in mini_quadrantes:
    if not mini_quadrantes[mini_quad]['selecionado']:
        nao_selecionados.append(mini_quad)
        
mini_quad = random.sample(nao_selecionados, 1)[0]
mini_quadrantes[mini_quad]['quantidade'] = 2
mini_quadrantes[mini_quad]['selecionado'] = True

values = mini_quadrantes[mini_quad]['dados'].values
values = np.unique(values)
vinte_dezenas.extend(random.sample(list(values), 2))
vinte_dezenas
```

### Montando Apostas

#### Montar Jogo a partir dos 20 numeros selecionados

- escolher entre 4-5 mini quadrantes (fazer isso random)
- selecionar 1 número aleatório de cada quadrante aleatório
- selecionar novamente até ter 6 números

```python
def montar_jogo(mini_quadrantes, vinte_dezenas, qtd_numeros):
    mini_quad_escolhidos = []
    for mini_quad in mini_quadrantes:
        if mini_quadrantes[mini_quad]['selecionado']:
            mini_quad_escolhidos.append(mini_quad)

    qtd_mini_quads = np.random.choice([4,5])
    mini_quad_escolhidos = random.sample(mini_quad_escolhidos, qtd_mini_quads)

    numeros_da_sorte = []
    for mini_quad in mini_quad_escolhidos:
        values = np.unique(mini_quadrantes[mini_quad]['dados'].values)
        values = list(set(values).intersection(vinte_dezenas))
        value = np.random.choice(values)
        numeros_da_sorte.append(value)
    
    valid_numbers = []
    values = []
    for mini_quad in mini_quad_escolhidos:
        values = np.unique(mini_quadrantes[mini_quad]['dados'].values)
        values = list(set(values).intersection(vinte_dezenas))
        valid_numbers.extend(list(values))


    while len(numeros_da_sorte) < qtd_numeros:
        value = np.random.choice(valid_numbers)

        if value not in numeros_da_sorte:
            numeros_da_sorte.append(value)
    numeros_da_sorte.sort()
    return numeros_da_sorte
```

```python
def montar_aposta(apostas):
    while True:
        print('A aposta vai ser de quantos números entre 6 e 15?')
        qtd_numeros = input()
        qtd_numeros = int(qtd_numeros)
        print()

        while True:
            if ((qtd_numeros > 15) | (qtd_numeros < 6)):
                print(
                    f'O número {qtd_numeros} está fora do range, digite um número entre 6 e 15')
                qtd_numeros = input()
                qtd_numeros = int(qtd_numeros)
                print()
            else:
                break

        print(f'Quantos jogos de {qtd_numeros} números você deseja gerar?')
        quantidade_desejada = input()
        quantidade_desejada = int(quantidade_desejada)
        print()

        for i in range(0,quantidade_desejada):
            aposta = montar_jogo(mini_quadrantes, vinte_dezenas, qtd_numeros)
            apostas.append(aposta)
            i+=1

        print('Deseja gerar mais jogos com outra quantidade de números?, se sim digite "s", caso não digite qualquer coisa')
        gerar_mais = input()
        print()

        if gerar_mais.lower() != 's':
            break
            
    return apostas
    
apostas = montar_aposta([])
```
